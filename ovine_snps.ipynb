```python
import pandas as pd
import numpy as np
import json
import shutup; shutup.please()
```

```python
raw_data = pd.read_csv('LLP_Genomed_Ayr_OVNG50V02_20230213_FinalReport.txt', sep='\t', header=0, skiprows=9)
```

```python
raw_data['Allele Pair'] = raw_data['Allele1 - Forward'] + raw_data['Allele2 - Forward']
raw_data.head()
```

```python
# number of SNPs
raw_data['SNP Name'].nunique()
```

```python
# number of animals
raw_data['Sample ID'].nunique()
```

```python
raw_data.shape
```

```python
data_table = raw_data.pivot_table(index = 'Sample ID', columns = 'SNP Name', values = 'Allele Pair', aggfunc=np.sum)
```

```python
# creating data table
data_table.head()
```

```python
data_table.shape
```

```python
data_json = {}
for index, row in data_table.iterrows():
    row_dict = {}
    for col, value in row.items():
        row_dict[col] = {"Allele1 - AB": value[0], "Allele2 - AB": value[1]}
    data_json[index] = row_dict
```

```python
len(data_json)
```

```python
data_json
```

```python
with open("ovine_SNPs.json", "w") as f:
    json.dump(data_json, f)
```

```python
data_json = []
for index, row in data_table.iterrows():
    row_dict = {'Sample ID': index}
    col_list = []
    for col, value in row.items():
        col_dict = {'SNP name': col, 'Allele1 - AB': value[0], 'Allele2 - AB': value[1]}
        col_list.append(col_dict)
    row_dict['SNPs'] = col_list
    data_json.append(row_dict)
```
